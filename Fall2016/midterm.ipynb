```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white", palette="muted", color_codes=True)
%matplotlib inline
```

```python
data = pd.read_csv('midterm.dat')
#data
```

```python
print('maximum score: {0:.2f}'.format(  data['midterm'].max() ) )
```

```python
print('minimum score: {0:.2f}'.format(  data['midterm'].min() ) )
```

```python
print('Average score: {0:.2f}'.format(  data['midterm'].mean() ) )
```

```python
ax = sns.boxplot(x=None, y="midterm", data=data)
#plt.title('APSC 3115 Midterm' )
```

```python
#rs = np.random.RandomState(10)

# Set up the matplotlib figure
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
sns.despine(left=True)

# Generate a random univariate dataset
#d = rs.normal(size=100)

# Plot a simple histogram with binsize determined automatically
sns.distplot(data, kde=False, color="b", ax=axes[0, 0])

# Plot a kernel density estimate and rug plot
sns.distplot(data, hist=False, rug=True, color="r", ax=axes[0, 1])

# Plot a filled kernel density estimate
sns.distplot(data, hist=False, color="g", kde_kws={"shade": True}, ax=axes[1, 0])

# Plot a historgram and kernel density estimate
sns.distplot(data, color="m", ax=axes[1, 1])

plt.setp(axes, yticks=[])
plt.tight_layout()
```
